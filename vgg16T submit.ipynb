```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adamax
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import numpy as np
import cv2
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16

# 얼굴 인식 모델 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def is_valid_image(face_image):
    aspect_ratio = float(face_image.shape[1]) / face_image.shape[0]
    if aspect_ratio < 0.4 or aspect_ratio > 2.5:
        return False
    mean_color = np.mean(face_image)
    if mean_color < 30 or mean_color > 225:
        return False
    if face_image.shape[0] < 100 or face_image.shape[1] < 100:
        return False
    return True

def preprocess_image(image):
    if image.dtype != np.uint8:
        image = (image * 255).astype(np.uint8)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_skin = np.array([0, 20, 70], dtype=np.uint8)
    upper_skin = np.array([20, 255, 255], dtype=np.uint8)
    mask = cv2.inRange(hsv, lower_skin, upper_skin)
    skin = cv2.bitwise_and(image, image, mask=mask)
    return skin

def load_images_and_labels(folder_path, label_mapping, img_size=(128, 128)):
    images = []
    labels = []
    class_counts = {class_name: 0 for class_name in label_mapping.keys()}

    for subfolder_name in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder_name)

        if os.path.isdir(subfolder_path) and subfolder_name in label_mapping:
            label = label_mapping[subfolder_name]

            for file_name in os.listdir(subfolder_path):
                file_path = os.path.join(subfolder_path, file_name)

                if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    try:
                        image = cv2.imread(file_path)
                        if image is None:
                            print(f"Warning: Image at {file_path} could not be loaded.")
                            continue

                        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                        # 얼굴 탐지 파라미터 조정
                        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

                        # 모든 탐지된 얼굴 처리
                        for (x, y, w, h) in faces:
                            face_image = image[y:y+h, x:x+w]
                            face_image = preprocess_image(face_image)  
                            face_image = cv2.resize(face_image, img_size)

                            if is_valid_image(face_image):
                                images.append(face_image / 255.0)  
                                labels.append(label)  # 클래스 인덱스를 추가합니다.
                                class_counts[subfolder_name] += 1

                    except Exception as e:
                        print(f"Error loading image {file_path}: {e}")

    print(f"Total images loaded: {len(images)}")  # 추가된 출력 메시지
    return np.array(images), np.array(labels), class_counts

def visualize_images(images, labels, label_mapping, num_images=5):
    plt.figure(figsize=(15, 10))
    labels = np.array(labels)  
    for i in range(len(label_mapping)):
        class_images = images[labels == i]
        if len(class_images) > 0:
            random_indices = np.random.choice(len(class_images), size=min(num_images, len(class_images)), replace=False)
            for j in range(len(random_indices)):
                plt.subplot(len(label_mapping), num_images, i * num_images + j + 1)
                plt.imshow(class_images[random_indices[j]])
                plt.axis('off')
                plt.title(list(label_mapping.keys())[i])
    plt.tight_layout()
    plt.show()

def build_model(input_shape, num_classes):
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)

    for layer in base_model.layers[:-4]:  
        layer.trainable = False

    inputs = layers.Input(shape=input_shape)
    x = base_model(inputs)
    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(256, activation='elu')(x)
    x = layers.Dropout(0.5)(x)  

    transformer_input = layers.Reshape((1, 256))(x)
    transformer_output = layers.MultiHeadAttention(num_heads=4, key_dim=256)(transformer_input, transformer_input)
    transformer_output = layers.LayerNormalization(epsilon=1e-6)(transformer_output + transformer_input)
    transformer_output = layers.Dense(256, activation='elu')(transformer_output)

    transformer_output = layers.GlobalAveragePooling1D()(transformer_output)

    outputs = layers.Dense(num_classes, activation='softmax')(transformer_output)

    model = models.Model(inputs=inputs, outputs=outputs)
    return model

def detect_faces(images):
    detected_images = []
    detected_labels = []
    for i, image in enumerate(images):
        image_uint8 = (image * 255).astype(np.uint8)  
        gray = cv2.cvtColor(image_uint8, cv2.COLOR_BGR2GRAY)
        # 얼굴 탐지 파라미터 조정
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
        if len(faces) > 0:
            detected_images.append(image)
            detected_labels.append(i)  
    return np.array(detected_images), np.array(detected_labels)


if __name__ == "__main__":
    train_folder_path = "/Users/withmocha/Desktop/DATA/BOAZ/미니 프로젝트 2/data/train face data"
    test_folder_path = "/Users/withmocha/Desktop/DATA/BOAZ/미니 프로젝트 2/data/train face data/test face data"

    label_mapping = {
        "spring": 0,
        "summer": 1,
        "fall": 2,
        "winter": 3
    }

    input_shape = (128, 128, 3)
    num_classes = len(label_mapping)

    # 이미지와 레이블 로드
    images, labels, class_counts = load_images_and_labels(folder_path=train_folder_path, label_mapping=label_mapping, img_size=(128, 128))

    encoder = OneHotEncoder(sparse_output=False)
    labels_onehot = encoder.fit_transform(labels.reshape(-1, 1))

    print("Counts of validated images per class:")
    for class_name, count in class_counts.items():
        print(f"Class '{class_name}': {count} images")

    # 얼굴 검수
    detected_images, detected_labels = detect_faces(images)
    detected_labels = labels[detected_labels]  # 검수된 이미지의 레이블

    print(f"Total detected images after face detection: {len(detected_images)}")

    # 각 클래스별 개수 계산
    unique, counts = np.unique(detected_labels, return_counts=True)
    class_counts_after_detection = dict(zip(unique, counts))

    print("Counts of detected images per class:")
    for class_index, count in class_counts_after_detection.items():
        class_name = list(label_mapping.keys())[class_index]
        print(f"Class '{class_name}': {count} images")

    # 검수된 이미지 시각화
    visualize_images(detected_images, detected_labels, label_mapping, num_images=5)

    # 데이터 증강
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    all_augmented_images = []
    all_augmented_labels = []

    for i in range(len(detected_images)):
        face_image = detected_images[i]
        label = detected_labels[i]

        all_augmented_images.append(face_image)
        all_augmented_labels.append(label)

        face_image = face_image.reshape((1,) + face_image.shape)  
        
        for _ in range(5):  # 10개의 증강 이미지를 생성
            augmented_image = next(datagen.flow(face_image, batch_size=1))
            all_augmented_images.append(augmented_image[0])  
            all_augmented_labels.append(label)

    all_augmented_images = np.array(all_augmented_images)
    all_augmented_labels = np.array(all_augmented_labels)

    print(f"Total augmented images: {len(all_augmented_images)}")  # 총 증강된 이미지 수 출력

    # 시각화
    visualize_images(all_augmented_images, all_augmented_labels, label_mapping, num_images=5)

    final_train_images = np.concatenate((detected_images, all_augmented_images), axis=0)
    final_train_labels = np.concatenate((encoder.transform(detected_labels.reshape(-1, 1)), encoder.transform(all_augmented_labels.reshape(-1, 1))), axis=0)

    final_counts = {class_name: 0 for class_name in label_mapping.keys()}
    for label in final_train_labels:
        class_index = np.argmax(label)
        class_name = list(label_mapping.keys())[class_index]
        final_counts[class_name] += 1

    print("Final counts of training images per class after augmentation:")
    for class_name, count in final_counts.items():
        print(f"Class '{class_name}': {count} images")

    # 모델 구축 및 훈련
    model = build_model(input_shape=input_shape, num_classes=num_classes)
    model.summary()
    
    optimizer = Adamax(learning_rate=0.0001)
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

    train_images, val_images, train_labels, val_labels = train_test_split(final_train_images, final_train_labels, test_size=0.2, random_state=42)

    history = model.fit(
        train_images,
        train_labels,
        validation_data=(val_images, val_labels),  
        epochs=100,
        callbacks=[early_stopping, reduce_lr],
        batch_size=16
    )

    # 모델 훈련 결과 시각화
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper right')

    plt.tight_layout()
    plt.show()

    # 테스트 데이터 로드
    test_images, test_labels, test_class_counts = load_images_and_labels(folder_path=test_folder_path, label_mapping=label_mapping, img_size=(128, 128))
    test_labels_encoded = encoder.transform(test_labels.reshape(-1, 1))
    test_predictions = model.predict(test_images)
    test_predictions_classes = np.argmax(test_predictions, axis=-1)
    test_labels_classes = np.argmax(test_labels_encoded, axis=1)

    # 예측 결과 출력
    print("\nTest Image Predictions:")
    for i, file_name in enumerate(os.listdir(test_folder_path)):
        if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            predicted_label = list(label_mapping.keys())[test_predictions_classes[i]]
            print(f"{file_name}: Predicted label = {predicted_label}")

    print(classification_report(test_labels_classes, test_predictions_classes, target_names=list(label_mapping.keys())))

    cm = confusion_matrix(test_labels_classes, test_predictions_classes)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(num_classes)
    plt.xticks(tick_marks, list(label_mapping.keys()), rotation=45)
    plt.yticks(tick_marks, list(label_mapping.keys()))
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()


    #input_image_path = input("Enter the path of the image to analyze: ")


    #input_image = cv2.imread(input_image_path)
    #if input_image is None:
        #print("Error loading image. Please check the file path.")
    #else:
        #input_image = cv2.resize(input_image, (128, 128)) / 255.0
        #input_image = np.expand_dims(input_image, axis=0)  # 배치 차원 추가

 
        #prediction = model.predict(input_image)
        #predicted_class_index = np.argmax(prediction, axis=-1)[0]
        #predicted_class = list(label_mapping.keys())[predicted_class_index]
        
        #print(f"The predicted color category for the input image is: {predicted_class}")
        
        
    model.save("trained_face_recognition_model.h5")
    print("Model saved as 'trained_face_recognition_model.h5'")
```
